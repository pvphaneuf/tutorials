```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generating a dataframe with binary values
df = pd.read_csv("data.csv", index_col=0)
df.head()
```

## Hierachical Clustering
Jaccard, Cosine and Correlation are good distance metrics for a matrix or binary values.

```python
import seaborn as sns
```

```python
distance_method_list = ["single", "complete", "average", "weighted", "centroid", "median", "ward"]
distance_metric_list = ["jaccard", "cosine", "correlation"]
```

```python
def output_clustermap(distance_method, distance_metric):
    print("distance_method:\t", distance_method)
    print("distance_metric:\t", distance_metric)
    sns.set(font_scale=1.2)
    g = sns.clustermap(df,
#                        figsize=(17, 17),
#                        linewidths=.5,
                       method=distance_method,
                       metric=distance_metric)
    g.cax.set_visible(False)
    plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(), rotation=0) # rotates vertical axis label
    plt.show()
```

```python
for distance_method in distance_method_list:
    for distance_metric in distance_metric_list:
        output_clustermap(distance_method, distance_metric)
```
